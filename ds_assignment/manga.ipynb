```python
import pandas as pd
import numpy as np
```

```python
#Initialize dataset
df = pd.read_csv("top500mangaMAL.csv")
```

```python
#Dropping useless columnds
df = df.drop(["Manga URL", "Type", "Volumes", "Chapters", "Status", "Published", "Published Dates", "Author",
 "Serialization", "Ranked", "Popularity", "Members", "Favorites", "Score Voted By", "Synonims Titles", "English Title"], axis=1)
df
```

```python
#List of all the genres
genres = df["Genres"][0]
genres
```

```python
#Look at ngrams in Genre
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
tfidf = vectorizer.fit_transform(df["Genres"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#Input
genre = "Adventure, Action, Fantasy, Demons, Drama"
#Term Frequency Inverse Document Frequency, converts text into numbers essentially and makes them unique values
query_vec = vectorizer.transform([genre])
#Find simlar values that match with the query
similarity = cosine_similarity(query_vec, tfidf).flatten()
#Get top 5 most similar to input
indices = np.argpartition(similarity, -5)[-5:]
#Input into the indices
results = df.iloc[indices]
results
```
